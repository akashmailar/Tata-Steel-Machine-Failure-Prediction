## **Project Summary -**

The **Tata Steel Machine Failure Prediction** project involves an **exploratory data analysis (EDA)** process applied to a dataset containing 136,430 rows and 14 columns. The dataset provides valuable information about the operational conditions of machines used in the manufacturing process, which can help predict machine failures. Early failure detection is critical for minimizing downtime, improving safety, and reducing maintenance costs. The project focuses on identifying key factors contributing to machine failures through a detailed examination of the dataset.

**Dataset Overview**

The dataset includes the following columns:

1. **Product ID** (String) – Unique identifier for each machine instance.
2. **Type** (String) – Categorical variable indicating the product type (e.g., L, M, H).
3. **Air Temperature [K]** (Float) – Temperature of the surrounding environment.
4. **Process Temperature [K]** (Float) – Internal process temperature of the machine.
5. **Rotational Speed [rpm]** (Integer) – Speed at which the machine operates.
6. **Torque [Nm]** (Float) – Rotational force applied to the machine.
7. **Tool Wear [min]** (Integer) – Cumulative wear on the machine’s tool.
8. **Machine Failure** (Boolean) – Target variable (0 = No Failure, 1 = Failure).
9. **TWF** (Tool Wear Failure) – Failure due to tool degradation.
10. **HDF** (Heat Dissipation Failure) – Failure caused by overheating.
11. **PWF** (Power Failure) – Failure due to power system issues.
12. **OSF** (Overstrain Failure) – Failure caused by excessive operational strain.
13. **RNF** (Random Failure) – Unclassified random failures
14. **id** (Interger) - Row Numbering.


**Data Preprocessing and Cleaning**

Before analyzing the data, several preprocessing steps are essential:

- **Handling Missing Values**: The dataset was inspected for missing values in any of the columns. Imputation methods (such as mean or median imputation) were used where necessary to fill in missing entries.

- **Data Type Conversion**: The columns were checked for correct data types, ensuring numerical columns were in appropriate formats for analysis, while categorical variables like 'Product ID' and 'Type' were kept as objects.

- **Outliers Detection**: Given the nature of industrial machinery, outliers may reflect abnormal operating conditions or sensor errors. Statistical methods like IQR (Interquartile Range) and Z-scores were used to detect and handle these outliers.


**Exploratory Data Analysis (EDA)**

1. **Distribution of Machine Failures**: One of the first steps in the analysis was to examine the distribution of the target variable, "Machine failure". A significant imbalance between failure and non-failure instances was observed, with the majority of machines not failing. This indicates the need for specialized techniques, such as resampling or class weighting, during modeling to account for this imbalance.

2. **Correlation Analysis**: A correlation matrix was created to examine the relationships between numerical features. Strong correlations were observed between some features, such as process temperature and air temperature, which could suggest a linear relationship. However, certain indices, such as TWF and RNF, showed weak correlations with each other and with the machine failure status, suggesting they might not be the strongest predictors for failure.

3. **Feature Importance**: Several features were found to have varying levels of importance in predicting machine failure. Features like rotational speed, torque, and tool wear were observed to exhibit higher importance. These features were likely indicative of the operational stress placed on the machine and its components.

4. **Class Imbalance**: The dataset showed a clear class imbalance, with significantly more instances of machines not failing compared to those that failed. This required strategies like SMOTE (Synthetic Minority Over-sampling Technique) to balance the dataset during model training.

5. **Categorical Feature Analysis**: The "Type" and "Product ID" columns were explored to see how different machines and types of products affected failure rates. Different types of machines showed different failure rates, with some types being more prone to failure under specific conditions.

6. **Time-Based Analysis**: Although there was no direct time feature, the tool wear data suggested temporal trends, with tool wear increasing over time correlating with higher failure rates. This insight can be used for predictive maintenance to monitor the condition of tools and replace them before failure occurs.


**Insights and Conclusions**
The primary goal of this EDA was to identify potential predictors for machine failure. Several important findings emerged:

- **Predictive Features**: Key features such as tool wear, rotational speed, torque, and process temperature were found to be highly predictive of machine failure. Monitoring these parameters can help anticipate failures and enable timely maintenance.

- **Class Imbalance**: The class imbalance requires special attention when building predictive models. Resampling techniques or algorithmic adjustments (such as class weights) would be necessary to train effective classifiers.

- **Operational Failure Indices**: Features like TWF, PWF, HDF, and RNF provide additional insights into the type of failure. These indices can be used to diagnose the specific cause of failure, aiding in more precise maintenance strategies.


In conclusion, this **EDA** highlights the importance of **continuous monitoring** and **predictive analytics** in industrial machinery. The insights gained will serve as the foundation for further model development and help guide **Tata Steel** in enhancing its **machine maintenance strategies**, **reducing unplanned downtime**, and **improving operational efficiency**.

### Variables Description

The dataset contains 14 columns (variables), each representing different aspects of machine performance and operational conditions. Below is the detailed description of each variable in the dataset:

1. **Product ID**
    - **Type**: Categorical (String)
    - **Description**: A unique identifier for each machine or product. This variable helps distinguish between different machines in the dataset.

2. **Type**
    - **Type**: Categorical (String)
    - **Description**: Indicates the type or category of the machine. Different machine types might have distinct operational characteristics and failure rates.

3. **Air temperature [K]**
    - **Type**: Continuous (Float)
    - **Description**: The temperature of the air surrounding the machine, measured in Kelvin. This temperature can influence machine performance and help predict potential failures.

4. **Process temperature [K]**
    - **Type**: Continuous (Float)
    - **Description**: The temperature inside the machine’s processing environment, measured in Kelvin. Elevated process temperatures might lead to overheating and impact the machine’s longevity.

5. **Rotational speed [rpm]**
    - **Type**: Continuous (Integer)
    - **Description**: The rotational speed of the machine, measured in revolutions per minute (rpm). Higher rotational speeds could put more stress on the machine components and increase the likelihood of failure.

6. **Torque [Nm]**
    - **Type**: Continuous (Float)
    - **Description**: The amount of torque applied to the machine, measured in Newton-meters (Nm). Excessive torque can indicate overworking or malfunctioning components, potentially leading to failure.

7. **Tool wear [min]**
    - **Type**: Continuous (Integer)
    - **Description**: The duration of tool usage in minutes before failure. As tools wear out over time, their effectiveness decreases, increasing the risk of failure.

8. **Machine failure**
    - **Type**: Categorical (Binary: 0 or 1)
    - **Description**: The target variable indicating whether a machine failure occurred (1 for failure, 0 for no failure). This is the variable we aim to predict using the other features.

9. **TWF (Tool Wear Failure)**
    - **Type**: Categorical (Binary: 0 or 1)
    - **Description**: The Tool Wear Failure, which tracks the failure of the tool due to wear. (1 for failure, 0 for no failure).

10. **HDF (Hydraulic Failure)**
    - **Type**: Categorical (Binary: 0 or 1)
    - **Description**: The Hydraulic Failure, which measures the performance of the hydraulic system. (1 for failure, 0 for no failure), which could lead to a machine malfunction.

11. **PWF (Power Failure)**
    - **Type**: Categorical (Binary: 0 or 1)
    - **Description**: The Power Failure, which monitors failures in the machine’s power system. (1 for failure, 0 for no failure), which could cause operational failures.

12. **OSF (Operational Failure)**
    - **Type**: Categorical (Binary: 0 or 1)
    - **Description**: The Operational Failure, which tracks other types of operational failures unrelated to power, hydraulics, or tool wear. (1 for failure, 0 for no failure), helps identify failures from other sources of operational stress.

13. **RNF (Rotational Failure)**
    - **Type**: Categorical (Binary: 0 or 1)
    - **Description**: The Rotational Failure, focusing on failures related to the rotational aspects of the machine. (1 for failure, 0 for no failure).


**Summary of Variable Types and Roles**:

- **Predictor Variables**: Air temperature, process temperature, rotational speed, torque, tool wear, TWF, HDF, PWF, OSF, RNF, and Type are all predictor variables. They provide information about the operational conditions and state of the machine, helping to predict whether a failure will occur.

- **Target Variable**: The "Machine failure" variable is the target variable (binary classification), where the goal is to predict whether the machine will fail (1) or not (0).

Each of these variables contributes important insights into the machine's condition, helping to develop a predictive maintenance model that can forecast machine failures and optimize maintenance scheduling.
